# Pruebas Gui vs Cli
## Introducción
En esta práctica se analizan las diferencias entre GUI (Graphical User Interface) y CLI (Command Line Interface) en el uso de Git y GitHub.

Además, se estudiaron conceptos fundamentales como commit, branch, pull request, fork, stash, remote, fetch y push.

# Conceptos básicos
## Commit

Un commit es un registro de cambios realizado en un proyecto.

git commit -m "Primer commit"

# Comparacion Gui vs Cli
| GUI | CLI |
|---|---|
| Fácil de usar | Más potente |
| Visual | Más rápido |
| Ideal para principiantes | Ideal para usuarios avanzados |

# Ejemplos 
## Cli
comandos_git = [
    "git init",
    "git add .",
    'git commit -m "Primer commit"',
    "git push origin main"
]

for comando in comandos_git:
    print(comando)

## Conclusión

Git permite gestionar versiones de proyectos de forma eficiente.
La GUI facilita el aprendizaje visual, mientras que la CLI proporciona mayor control y velocidad.
